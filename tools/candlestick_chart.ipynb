# Notebook para gráfico de velas
Este notebook genera un gráfico de velas de precios con estilo elegante para el dashboard.

```python
# 1. Importar librerías
import pandas as pd
import mplfinance as mpf
import os

print('Librerías cargadas')
```

```python
# 2. Cargar datos de precios
# Ajusta esta ruta al CSV con datos históricos de precio.
file_path = os.path.join('data', 'prices.csv')

if not os.path.exists(file_path):
    raise FileNotFoundError(f'No se encontró el archivo de datos: {file_path}')

prices = pd.read_csv(file_path)
prices.head()
```

```python
# 3. Preparar datos para el gráfico de velas
prices['Date'] = pd.to_datetime(prices['Date'])
prices.set_index('Date', inplace=True)

required = {'Open', 'High', 'Low', 'Close', 'Volume'}
if not required.issubset(prices.columns):
    raise ValueError(f'El DataFrame debe contener las columnas: {required}')

prices = prices[['Open', 'High', 'Low', 'Close', 'Volume']]
prices.tail()
```

```python
# 4. Crear gráfico de velas con estilo elegante
mc = mpf.make_marketcolors(up='#22c55e', down='#ef4444', wick='white', edge='white', volume='in')
s  = mpf.make_mpf_style(base_mpf_style='nightclouds', marketcolors=mc,
                        gridstyle='-.', gridcolor='#334155', facecolor='#0a0e27', edgecolor='#0f172a')

mpf.plot(prices,
         type='candle',
         style=s,
         title='Gráfico de Velas - Precio Histórico',
         ylabel='Precio',
         volume=True,
         mav=(9, 21),
         figsize=(12, 8),
         tight_layout=True)
```

```python
# 5. Añadir indicadores técnicos
prices['SMA_20'] = prices['Close'].rolling(window=20).mean()
prices['SMA_50'] = prices['Close'].rolling(window=50).mean()

apds = [
    mpf.make_addplot(prices['SMA_20'], color='#38bdf8', width=1.5, linestyle='-'),
    mpf.make_addplot(prices['SMA_50'], color='#facc15', width=1.5, linestyle='--')
]

mpf.plot(prices,
         type='candle',
         style=s,
         addplot=apds,
         title='Gráfico de Velas con Indicadores SMA20/SMA50',
         ylabel='Precio',
         volume=True,
         figsize=(12, 8),
         tight_layout=True)
```

```python
# 6. Mostrar y exportar el gráfico
output_file = os.path.join('release', 'candlestick_chart.png')
mpf.plot(prices,
         type='candle',
         style=s,
         addplot=apds,
         title='Gráfico de Velas con Indicadores SMA20/SMA50',
         ylabel='Precio',
         volume=True,
         figsize=(12, 8),
         tight_layout=True,
         savefig=output_file)

print(f'Gráfico exportado a: {output_file}')
```
